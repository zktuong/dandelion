```python
import numpy as np
from scipy.sparse import csr_matrix
import os
import scanpy as sc
import dandelion as ddl

# change directory to somewhere more workable

os.chdir(os.path.expanduser("~/Downloads/dandelion_tutorial/"))
sc.settings.verbosity = 3

ddl.logging.print_versions()
```

```python
# read in the airr_rearrangement.tsv file
file_location = (
    "sc5p_v2_hs_PBMC_10k/sc5p_v2_hs_PBMC_10k_t_airr_rearrangement.tsv"
)

# read in gene expression data
adata = sc.read_10x_h5(
    "sc5p_v2_hs_PBMC_10k/sc5p_v2_hs_PBMC_10k_filtered_feature_bc_matrix.h5"
)
adata.var_names_make_unique()
adata
```

```python
vdj = ddl.read_10x_airr(file_location)
vdj
```

```python
vdj, adata = ddl.pp.check_contigs(
    vdj, adata, ntop_vdj=1, ntop_vj=1, filter_ambiguous=True
)
# sample to 1M cells to test the full pipeline
vdj, adata = ddl.tl.vdj_sample(vdj, adata, size=1_000_000, random_state=42)
```

```python
ddl.tl.find_clones(vdj)
ddl.tl.clone_size(vdj)
vdj.metadata["clone_id_size_category"].value_counts()
```

```python
ddl.tl.generate_network(
    vdj,
    key="junction",
    distance_mode="full",
    lazy=True,
    return_graph=False,
    num_cores=8,
)
vdj
```
