## Test Column Assignment

Now test adding new columns to the data


```python
# Test 1: Add a constant column
vdj.data["test_constant"] = 42
print("Added constant column")
print(vdj.data.select(["sequence_id", "test_constant"]).head(3).collect())
```

```python
# Test 2: Add a column from an expression
vdj.data["v_gene"] = pl.col("v_call").str.extract(r"(IGH\w*)", 0)
print("Added column from expression")
print(vdj.data.select(["v_call", "v_gene"]).head(3).collect())
```

```python
# Test 3: Add a column from arithmetic operations
vdj.data["consensus_plus_umi"] = pl.col("consensus_count") + pl.col("umi_count")
print("Added column from arithmetic operation")
print(
    vdj.data.select(["consensus_count", "umi_count", "consensus_plus_umi"])
    .head(3)
    .collect()
)
```

```python
# Test 4: Add a column to metadata
vdj.metadata["test_metadata_col"] = "test_value"
print("Added column to metadata")
print(vdj.metadata.select(["cell_id", "test_metadata_col"]).head(3).collect())
```

## Summary: Column Assignment Support

The `DataFrameAccessor` now supports pandas-like column assignment for both `.data` and `.metadata`:

### Supported Assignment Types:
1. **Constant values**: `vdj.data["col"] = 42`
2. **Lists/tuples**: `vdj.data["col"] = [1, 2, 3, ...]`
3. **Series**: `vdj.data["col"] = some_series`
4. **Expressions**: `vdj.data["col"] = pl.col("other_col") + 10`
5. **Arithmetic operations**: `vdj.data["col"] = pl.col("col1") + pl.col("col2")`

Changes made to `.data` automatically update the underlying `vdj._data`, and changes to `.metadata` update `vdj._metadata`.
